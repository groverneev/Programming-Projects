```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os

def get_links(url):
    try:
        response = requests.get(url, timeout=5)
        soup = BeautifulSoup(response.content, 'html.parser')
        links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]
        return links, response.text  # Return both links and page content
    except Exception as e:
        print(f"Failed to fetch {url}: {e}")
        return [], ""

def crawl(url, depth, output_dir="crawled_pages"):
    os.makedirs(output_dir, exist_ok=True)  # Create directory if not exists

    crawled_urls = set()
    urls_to_crawl = [(url, 0)]

    while urls_to_crawl:
        current_url, current_depth = urls_to_crawl.pop(0)
        
        if current_url not in crawled_urls and current_depth <= depth:
            print(f"Crawling {current_url}")
            crawled_urls.add(current_url)

            links, page_content = get_links(current_url)

            if page_content:
                # Save the page content
                filename = os.path.join(output_dir, f"page_{len(crawled_urls)}.html")
                with open(filename, "w", encoding="utf-8") as file:
                    file.write(page_content)

            for link in links:
                urls_to_crawl.append((link, current_depth + 1))

if __name__ == "__main__":
    start_url = input("Enter the URL to start crawling: ")
    crawl_depth = int(input("Enter the depth of crawling: "))
    crawl(start_url, crawl_depth)
```
