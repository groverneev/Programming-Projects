```python
import tkinter as tk
import chess
import chess.svg
from stockfish import Stockfish

# Set up Stockfish
stockfish_path = "/opt/homebrew/bin/stockfish"  # Path to your Stockfish binary
stockfish = Stockfish(stockfish_path)

# Set up the board
board = chess.Board()

# Create the Tkinter window
window = tk.Tk()
window.title("Chess with Stockfish")

# Create a canvas to draw the chessboard
canvas = tk.Canvas(window, width=400, height=400)
canvas.pack()

# Define colors
light_square = "#f0d9b5"
dark_square = "#b58863"

# Draw the chessboard
def draw_board():
    for row in range(8):
        for col in range(8):
            x1 = col * 50
            y1 = row * 50
            x2 = x1 + 50
            y2 = y1 + 50
            color = light_square if (row + col) % 2 == 0 else dark_square
            canvas.create_rectangle(x1, y1, x2, y2, fill=color)
            
# Place the pieces on the board
def draw_pieces():
    for square in chess.SQUARES:
        piece = board.piece_at(square)
        if piece:
            row, col = divmod(square, 8)
            piece_image = get_piece_image(piece.symbol())
            canvas.create_image(col * 50 + 25, row * 50 + 25, image=piece_image)

# Load image for each piece
def get_piece_image(piece_symbol):
    piece_images = {
        'P': '♙', 'N': '♘', 'B': '♗', 'R': '♖', 'Q': '♕', 'K': '♔', 
        'p': '♟', 'n': '♞', 'b': '♝', 'r': '♜', 'q': '♛', 'k': '♚'
    }
    return piece_images.get(piece_symbol, '')

# Function to update the board
def update_board():
    canvas.delete("all")
    draw_board()
    draw_pieces()

# Function to make a move and get the best move from Stockfish
def make_move():
    move = stockfish.get_best_move()
    board.push(chess.Move.from_uci(move))
    update_board()

# Button to make the best move
move_button = tk.Button(window, text="Stockfish Move", command=make_move)
move_button.pack()

# Initial draw
update_board()

# Run the Tkinter main loop
window.mainloop()
```
